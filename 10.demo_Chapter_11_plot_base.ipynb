```r
library('plotrix')
```

```r
x <- c(46,37,35,53)
labels <- c('10A1','10A2','10A3','10A4')
# png(file='classes_pie_my.jpg')
pie(x,labels)
dev.off()
```

```r
# png(file='classes_pie_my.jpg')
pie(x, labels, main='CLasses pie chart', 
    col=rainbow(length(x)),
   clockwise=TRUE)
dev.off()
```

```r
# example(pie)
```

```r
# png(file='classes_pie_my.jpg')
piepercent <- round(100*x/sum(x),2)
pie(x, labels=paste(labels,'-',piepercent,'%'), main='CLasses pie chart', 
    col=rainbow(length(x)),
   clockwise=TRUE)
legend('topright',labels,cex=0.5,fill=rainbow(length(x)))
dev.off()
```

```r
piepercent <- round(100*x/sum(x),2)
pie3D(x,labels=paste(labels,'-',piepercent,'%'), main='Classes Pie Chart',
     col=rainbow(length(x)),explode=0.09)
legend('topright',labels,cex=0.5,fill=rainbow(length(x)))
dev.off()
```

```r
H <- c(5,15,20,5)
M <- c('Excellent','Very Good','Good','Midium')

# png(file='my_bar.png')

barplot(H, names.arg=M, xlab='Grade',ylab='Students',
       col='#348feb', main='Grade of class', border='#4634eb')
dev.off()
```

```r
colors <- c('#749c6a','#eb6e34','#7d8085','#348feb')
classes <- c('10A1','10A2','10A3')
grades <- c('Excellent','Very Good','Good','Medium')

values <- matrix(c(50,6,7,15,14,13,20,21,22,5,40,3),
                nrow=4, ncol=3, byrow=TRUE)
print(values)
```

```r
barplot(values, main='Grade of 3 classes', names.arg=classes,
       xlab='Class', ylab='Students',col=colors)
legend('top', grades, cex=0.9, fill=colors,border='#ecede6')

dev.off()
```

```r
barplot(values, main='Grade of 3 classes', names.arg=classes,
       xlab='Class', ylab='Students',col=colors)
legend('topright', grades, cex=0.9, fill=colors,border='#ecede6',inset=0.1)
dev.off()
```

```r
bears <- c(8,54,93,116,137)
plot(bears, type='o', col='blue',xlab='Year',ylab='Numbers',
    main='Number of bears chart')
axis(1,at=1:6, labels=c('2017','2018','2019','2020','2021','2022'))
dev.off()
```

```r
bears <- c(8,54,93,116,137)
dolphins <- c(150,77,32,11,6,1)
whales <- c(80,54,100,76,93,72)

g_range <- range(0,bears,dolphins,whales)
plot(bears, type='o', col='red',xlab='Year',ylab='Numbers',
    main='Number of bears chart',xaxt='n')

box()
lines(dolphins, type='o',col='blue')
lines(whales, type='o',col='green')

legend('topleft', c('bears','dolphins','whales'), cex=0.8,
      fill=c('red','blue','green'))
axis(1,at=1:6, labels=2017:2022)
dev.off()
```
