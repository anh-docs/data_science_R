## Normal Distribution

```r
# dnorm(x, mean, sd) : Mat do
```

```r
x <- seq(-10,10,0.1)
```

```r
y <- dnorm(x,mean=0,sd=1)
```

```r
plot(x,y)
```

```r
# qnorm(p, mean, sd): Dinh bac
```

```r
x <- seq(0,1, by=0.02)
```

## Binomial Diatribution

```r
# dbinom(x, size, prob)
x<- seq(0, 50, by=1)
```

```r
y <- dbinom(x,50,0.5)
```

```r
plot(x,y)
```

```r
dbinom(2,3,0.6)
```

```r
# pbinom(x, sixe, prob)
p_x_25_less <- pbinom(25,50,0.5)
```

```r
p_x_25_less
```

```r
p_x_25_more <- 1- pbinom(24,50,0.5)
```

```r
p_x_25_more
```

```r
# qbinom(p, sixe, prob)
x <- qbinom(0.25,50,0.5)
```

```r
x
```

```r
# rbinom(n, size, prob)
x <- rbinom(8,100,1/6)
```

```r
x
```
