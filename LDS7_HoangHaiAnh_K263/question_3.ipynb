```r
#1.
vector <- c(12.2,38.9,75.4,69.0,158.0,161.7,119.9,140.9,89.4,134.9,68.8,158.7,
119.2,31.9,14.9,13.5,181.5,120.3,133.0,107.6,137.7,134.6,86.3,87.5,
98.9,43.8,32.4,114.3,204.7,178.0,124.0,157.7,101.0,139.0,83.6,44.6,
39.6,91.7,22.7,64.6,143.5,192.8,152.4,129.4,119.4,144.5,104.2,135.0,
49.7,72.9,45.6,81.7,147.9,123.9,111.6,107.6,97.9,93.7,75.1,67.6)
```

```r
#2.
print(vector)
```

```r
#4.
data.timeseries <- ts(vector, start=c(2013,1), frequency=12)
```

```r
data.timeseries
```

```r
plot(data.timeseries, main='Timeseries by Sunny in HaNoi',
    col='red', xlab='Year', ylab='hours')
dev.off()
```

- Nắng tại Hà Nội biến động rất mạnh
- Đỉnh điểm rơi vào giữa năm 2015

```r
comp <- decompose(data.timeseries)
plot(comp)
```

- trend: số giờ nắng theo năm biến động rất mạnh, đạt đỉnh tại giữa năm 2015 và cuối năm 2016
giữa năm 2017 số giờ nắng đã chạm đáy
- seasonal: số giờ nắng khá chịu ảnh hưởng theo mùa 
- random: số giờ nắng cũng không đều

```r
#5.
data.timeseries.forecasts <- HoltWinters(data.timeseries)
print(data.timeseries.forecasts)
```

- alpha: 3% giá trị hiện tại không phụ thuộc vào các dữ liệu quá khứ gần
- beta: ~ 0 độ dốc của trend không cập nhật trong thời gian qua
- gamma: 54% có sự ảnh hưởng theo mùa 

```r
plot(data.timeseries.forecasts, main='observed & forecasts',
    xlab='Year', ylab='Hours',col=c('blue','red'))
dev.off()
```

- Kết quả dự báo khá gần quan sát

```r
data.timeseries.forecasts_new <- forecast:::forecast.HoltWinters(data.timeseries.forecasts, h=6)
```

```r
data.timeseries.forecasts_new
```

```r
plot(data.timeseries.forecasts_new, main='Sunny: forecasts next 6 month',
    xlab='Year', ylab='Hours',col=c('blue','red'))
dev.off()
```
