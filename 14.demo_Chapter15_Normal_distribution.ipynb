```r
x <- seq(-10,10,by=0.1)
y <- dnorm(x, mean=0, sd=1)
plot(x,y)
```

```r
x <- seq(-10,10,by=0.1)
y <- pnorm(x, mean=0, sd=1)
plot(x,y)
```

```r
x <- seq(0,1, by=0.02)
y <- qnorm(x, mean=0, sd=1)
plot(x,y)
```

```r
y <- rnorm(50)
y
```

# Binomial Distribution

```r
p1 <- dbinom(2,3,0.6)
p1
```

```r
p_x_25_less <- pbinom(25,50,0.5)
p_x_25_less
```

```r
p_x_25_more <- 1 - pbinom(25,50,0.5)
p_x_25_more
```

```r
x <- qbinom(0.25,50, 0.5)
x
```

```r
x <- rbinom(8,100,1/6)
print(x)
```
